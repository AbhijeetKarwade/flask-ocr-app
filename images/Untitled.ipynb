```python
import cv2
import matplotlib.pyplot as plt
import pytesseract

# Load the image
image_path = "receipt001.jpg"
image = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Crop the receipt portion (manual estimation based on image content)
height, width = gray_image.shape
cropped_image = gray_image[int(height * 0.25):int(height * 0.8), int(width * 0.2):int(width * .92)]

# Display the processed images
plt.figure(figsize=(10, 5))

# Grayscale image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

# Cropped image
plt.subplot(1, 2, 2)
plt.imshow(cropped_image, cmap='gray')
plt.title("Cropped Image")
plt.axis("off")

plt.show()

# Use pytesseract to extract text from the cropped grayscale image
extracted_text = pytesseract.image_to_string(cropped_image)

# Display the extracted text
print("Extracted Text:")
print(extracted_text)
```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image

# Load the uploaded image
image_path = "receipt011.jpg"
image = cv2.imread(image_path)

# Convert to grayscale for better OCR
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Adjust brightness and contrast
alpha = 1.5  # Contrast control
beta = 50     # Brightness control
bright_image = cv2.convertScaleAbs(gray_image, alpha=alpha, beta=beta)

# Apply OCR using Tesseract
custom_config = r'--oem 3 --psm 6'  # Use LSTM OCR Engine, assume block of text
extracted_text = pytesseract.image_to_string(bright_image, config=custom_config)

# Display extracted text
extracted_text
```

```python
import cv2
import matplotlib.pyplot as plt
import pytesseract

# Load the image
image_path = "receipt021.jpg"
image = cv2.imread(image_path)

# Check if the image is loaded correctly
if image is None:
    print("Error: Unable to load the image. Check the file path.")
    exit()

# Increase resolution by scaling up the image
scale_factor = 2  # Increase resolution by 2x
high_res_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_CUBIC)

# Convert to grayscale
gray_image = cv2.cvtColor(high_res_image, cv2.COLOR_BGR2GRAY)

# Apply bilateral filtering for noise reduction while preserving edges
filtered_image = cv2.bilateralFilter(gray_image, 9, 75, 75)

# Crop the receipt portion (manual estimation based on image content)
height, width = filtered_image.shape
cropped_image = high_res_image[int(height * 0.0):int(height * 0.9), int(width * 0.1):int(width * .9)]

# Display the processed images
plt.figure(figsize=(12, 6))

# High-resolution grayscale image
plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("High-Resolution Grayscale")
plt.axis("off")

# Filtered image
plt.subplot(1, 3, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title("Filtered Image")
plt.axis("off")

# Cropped image
plt.subplot(1, 3, 3)
plt.imshow(cropped_image, cmap='gray')
plt.title("Cropped Image")
plt.axis("off")

plt.show()

# Use pytesseract to extract text from the cropped high-resolution grayscale image
custom_config = r'--oem 3 --psm 6 -c tessedit_char_whitelist="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789$.:/"'
extracted_text = pytesseract.image_to_string(cropped_image, config=custom_config, lang='eng')

# Display the extracted text
print("Extracted Text:")
print(extracted_text)
```
